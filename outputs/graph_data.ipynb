```python
import matplotlib.pyplot as plt          
import os
import pandas as pd

# Load Data
training_losses = pd.read_csv('./training_losses.csv');
validation_losses = pd.read_csv('./validation_losses.csv');
bleu_scores = pd.read_csv('./bleu_scores.csv');
```

```python
# Plot Training Losses
fig1 = plt.figure()
fig1.suptitle('Training Loss vs Minibatch', fontsize=20)
plt.xlabel('minibatch', fontsize=16)
plt.ylabel('loss', fontsize=16)
training_losses.iloc[:,2].plot()

# Plot Validation Losses
fig2 = plt.figure()
fig2.suptitle('Validation Loss vs Minibatch', fontsize=20)
plt.xlabel('minibatch', fontsize=16)
plt.ylabel('loss', fontsize=16)
validation_losses.iloc[:,2].plot()

# Plot Bleu Scores
fig3 = plt.figure()
fig3.suptitle('Bleu Scores vs Minibatch', fontsize=20)
plt.xlabel('minibatch', fontsize=16)
plt.ylabel('loss', fontsize=16)
bleu_scores.iloc[:,2].plot()
```

```python
cnntype = "Deep"

losses = pd.read_csv('./results/kfold-deep-full/trace-0');

# average losses
fig1 = plt.figure()
fig1.suptitle('Training Loss for ' + cnntype + ' CNN', fontsize=20)
plt.xlabel('minibatch * 50', fontsize=16)
plt.ylabel('loss', fontsize=16)
losses.iloc[:,2].plot()

plt.savefig(cnntype + " training loss")
plt.figure()

losses = pd.read_csv('./results/kfold-deep-full/val-agg-0');
# accuracy
fig1 = plt.figure()
fig1.suptitle('Accuracy for ' + cnntype + ' CNN', fontsize=20)
plt.xlabel('minibatch * 300', fontsize=16)
plt.ylabel('accuracy', fontsize=16)
#plt.ylim(0, 100)
losses.iloc[:,0].plot()
plt.savefig(cnntype + " accuracy")
plt.figure()

losses = pd.read_csv('./results/kfold-deep-full/val-loss-0');
# validation loss
fig1 = plt.figure()
fig1.suptitle('Validation Loss for ' + cnntype + ' CNN', fontsize=20)
plt.xlabel('minibatch * 300', fontsize=16)
plt.ylabel('loss', fontsize=16)
#plt.ylim(0, 100)
losses.iloc[:,1].plot()
plt.savefig(cnntype + " validation loss")
plt.figure()
```
