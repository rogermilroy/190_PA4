```python
import matplotlib.pyplot as plt          
import os
import pandas as pd

# Load LSTM Data
training_losses_lstm = pd.read_csv('./training_losses_lstm.csv');
validation_losses_lstm = pd.read_csv('./validation_losses_lstm.csv');
bleu_scores_lstm = pd.read_csv('./bleu_scores_lstm.csv');

# Load GRU data
training_losses_gru = pd.read_csv('./training_losses_gru.csv');
validation_losses_gru = pd.read_csv('./validation_losses_gru.csv');
bleu_scores_gru = pd.read_csv('./bleu_scores_gru.csv');
```

```python
print(training_losses_lstm)
```

```python
# Plot Training Losses
fig1 = plt.figure()
fig1.suptitle('LSTM Training Loss vs Minibatch', fontsize=20)
plt.xlabel('epoch/2', fontsize=16)
plt.ylabel('loss', fontsize=16)
training_losses_lstm.iloc[:,1].plot()
fig1.savefig('training_loss_lstm.png')

# Plot Validation Losses
fig2 = plt.figure()
fig2.suptitle('LSTM Validation Loss vs Minibatch', fontsize=20)
plt.xlabel('epoch/2', fontsize=16)
plt.ylabel('loss', fontsize=16)
validation_losses_lstm.iloc[:,1].plot()
fig2.savefig('validation_loss_lstm.png')

# Plot Bleu Scores
fig3 = plt.figure()
fig3.suptitle('LSTM Bleu Scores vs Minibatch', fontsize=20)
plt.xlabel('epoch/2', fontsize=16)
plt.ylabel('loss', fontsize=16)
bleu_scores_lstm.iloc[:,1].plot()
fig3.savefig('bleu_scores_lstm.png')
```

```python
# Plot Training Losses
fig1 = plt.figure()
fig1.suptitle('GRU Training Loss vs Minibatch', fontsize=20)
plt.xlabel('epoch/2', fontsize=16)
plt.ylabel('loss', fontsize=16)
training_losses_gru.iloc[:,1].plot()
fig1.savefig('training_loss_gru.png')

# Plot Validation Losses
fig2 = plt.figure()
fig2.suptitle('GRU Validation Loss vs Minibatch', fontsize=20)
plt.xlabel('epoch/2', fontsize=16)
plt.ylabel('loss', fontsize=16)
validation_losses_gru.iloc[:,1].plot()
fig2.savefig('validation_loss_gru.png')

# Plot Bleu Scores
fig3 = plt.figure()
fig3.suptitle('GRU Bleu Scores vs Minibatch', fontsize=20)
plt.xlabel('epoch/2', fontsize=16)
plt.ylabel('loss', fontsize=16)
bleu_scores_gru.iloc[:,1].plot()
fig3.savefig('bleu_scores_gru.png')
```
